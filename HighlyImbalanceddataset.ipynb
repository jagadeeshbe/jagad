Imbalanced Dataset:
A dataset with unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data['Class'].value_counts()
```

This is Highly Imabalanced Dataset

0 -- > Legit transaction
1 -- > Fraudulent transaction

```python
# seperating the legit and fraudulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Undersampling

Build a sample dataset containing similar distribution of legit & Fraud transactions

```python
legit_sample= legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
