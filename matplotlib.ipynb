Useful for making plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x,y,z)
```

```python
# plotting the data
```

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title, x axis, y axis
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.show()
```

```python
# parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g--')
```

```python
plt.plot(x,y,'bo')
```

```python
plt.plot(x,y,'y.')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Latin','German']
people=[100,200,300,400]
ax.bar(languages,people)
plt.show()
```

pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Latin','German']
people=[100,200,300,400]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3d scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z= 20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
