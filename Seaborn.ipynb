Data Visualization library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Seaborn has some builtin data sets

```python
# total bill vs tips
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load the iris
iris=sns.load_dataset('iris')
iris.head()
```

Scatter plot

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species',style='species')
```

```python
# loading the titanic data set
titanic = sns.load_dataset('titanic')
titanic.head()
```

count plot

```python
sns.countplot(data=titanic,x='class')
```

```python
titanic.shape
```

```python
sns.countplot(data=titanic,x='survived')
```

Bar chart

```python
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```

```python
# house price data set
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
housing= pd.DataFrame(housing.data,columns=housing.feature_names)
housing[' HouseAge'] = housing.pop('HouseAge')
housing.head()
```

Distribution Plot

```python
sns.displot(data=housing,x=' HouseAge',kde=True)
```

Correlation
+ correlation
- correlation

Heat map

```python
plt.figure(figsize=(10,10))
sns.heatmap(housing.corr(),annot=True,cmap='coolwarm')
```
